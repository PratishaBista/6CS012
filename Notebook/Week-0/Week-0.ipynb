```python
print("Categories:")
print("l : Length (m, ft)")
print("w : Weight (kg, lbs)")
print("v : Volume (L, gal)\n")

cat = input("Choose category (l/w/v): ")

if cat == "l":
    print("Available units: m (meters), ft (feet)")
elif cat == "w":
    print("Available units: kg (kilograms), lbs (pounds)")
elif cat == "v":
    print("Available units: L (liters), gal (gallons)")
else:
    print("Invalid category")
    exit()

unit1 = input("Convert from: ")
unit2 = input("Convert to: ")

try:
    num1 = float(input("Enter value: "))
except ValueError:
    print("Invalid number entered")
    exit()

def convert(unit1, unit2, num1):
    """
    Converts the entered value from one unit to another.

    Parameters:
    unit1 (str): the unit to convert from
    unit2 (str): the unit to convert to
    num1 (float): the value to be converted

    Returns:
    float or str: converted value if valid otherwise throws error message
    """

    # length
    if unit1 == "m" and unit2 == "ft":
        return num1 * 3.28084
    elif unit1 == "ft" and unit2 == "m":
        return num1 / 3.28084

    # weight
    elif unit1 == "kg" and unit2 == "lbs":
        return num1 * 2.204623
    elif unit1 == "lbs" and unit2 == "kg":
        return num1 / 2.204623

    # volume
    elif unit1 == "L" and unit2 == "gal":
        return num1 * 0.2641720524
    elif unit1 == "gal" and unit2 == "L":
        return num1 / 0.2641720524

    else:
        return "Invalid conversion"

ans = convert(unit1, unit2, num1)

print(num1, unit1, "in", unit2, "is", ans)
```

```python
operations = {
    "1": "Sum",
    "2": "Average",
    "3": "Maximum",
    "4": "Minimum"
}

print("Available operations:")
for key, name in operations.items():
    print(f"{key}: {name}")
print()

while True:
    cat = input("Choose operation (1/2/3/4): ")

    if cat not in operations:
        print("Invalid operation choice. Please enter 1, 2, 3, or 4.\n")
        continue  

    nums = input("Enter numbers separated by spaces: ")

    try:
        num_list = [float(n) for n in nums.split()]
        if len(num_list) == 0:
            print("List is empty. Try again.\n")
            continue  
    except ValueError:
        print("Invalid input: enter only numbers.\n")
        continue  

    break


def operate(cat, num_list):
    """
    Performs a mathematical operation on a list of numbers.

    Parameters:
    cat (str): operation type ("1"=sum, "2"=average, "3"=max, "4"=min)
    num_list (list): list of numeric values

    Returns:
    float: result of operation
    """

    if cat == "1":  # sum
        total = 0
        for n in num_list:
            total += n
        return total

    elif cat == "2":  # average
        total = 0
        for n in num_list:
            total += n
        return total / len(num_list)

    elif cat == "3":  # maximum
        maximum = num_list[0]
        for n in num_list:
            if n > maximum:
                maximum = n
        return maximum

    elif cat == "4":  # minimum
        minimum = num_list[0]
        for n in num_list:
            if n < minimum:
                minimum = n
        return minimum


ans = operate(cat, num_list)

print(f"\nThe {operations[cat]} of the numbers {num_list} is {ans}")
```

## 4.3. Exercise on nested list

1. Flatten a nested list

```python
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
    lst (list): nested list (list of lists)

    Returns:
    list: flattened list with all elements in a single dimension
    """
    result = []
    for sublist in lst:        
        for item in sublist:   
            result.append(item)
    return result

nested_list = [[1, 2], [3, 4], [5]]
flat_list = flatten(nested_list)

print("Original nested list:", nested_list)
print("Flattened list:", flat_list)
```

2. Accessing nested list elements

```python
def access_nested_element(lst, indices):
    """
    Returns the element from a nested list using a list of indices.

    Parameters:
    lst (list): nested list
    indices (list): list of indices to navigate the nested list

    Returns:
    element: the element at the specified nested position
    """
    result = lst
    for index in indices:
        result = result[index]  
    return result

nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 0]  

element = access_nested_element(nested_list, indices)

print("Nested list:", nested_list)
print(f"Element at indices {indices}:", element)
```

3. Sum of All Elements in a Nested List

```python
def sum_nested(lst):
    """
    Recursively calculates the sum of all numbers in a nested list.

    Parameters:
    lst (list): nested list of numbers (can contain sublists)

    Returns:
    int/float: sum of all numbers in the nested list
    """
    total = 0
    for item in lst:
        if isinstance(item, list):      
            total += sum_nested(item)
        else:                            
            total += item
    return total

nested_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(nested_list)

print("Nested list:", nested_list)
print("Sum of all elements:", result)
```

4. Remove Specific Element from a Nested List

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
    lst (list): nested list (list of lists)
    elem: element to remove

    Returns:
    list: nested list with the element removed from all sublists
    """
    result = []
    for sublist in lst:
        new_sublist = [x for x in sublist if x != elem]
        result.append(new_sublist)
    return result

nested_list = [[1, 2], [3, 2], [4, 5]]
element_to_remove = 2

modified_list = remove_element(nested_list, element_to_remove)

print("Original nested list:", nested_list)
print(f"Nested list after removing {element_to_remove}:", modified_list)
```

5. Find the Maximum Element in a Nested List

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list (any depth).

    Parameters:
    lst (list): nested list

    Returns:
    int/float: maximum element
    """
    maximum = float('-inf')
    for item in lst:
        if isinstance(item, list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum, item)
    return maximum

nested_list = [[1, 2], [3, [4, 5]], 6]
print("Maximum:", find_max(nested_list))
```

6. Count Occurrences of an Element in a Nested List

```python
def count_occurrences(lst, elem):
    """
    Counts how many times elem appears in a nested list.

    Parameters:
    lst (list): nested list
    elem: element to count

    Returns:
    int: number of occurrences
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

nested_list = [[1, 2], [2, 3], [2, 4]]
print("Occurrences of 2:", count_occurrences(nested_list, 2))
```

7. Flatten a List of Lists of Lists

```python
def deep_flatten(lst):
    """
    Flattens a nested list of arbitrary depth into a single list.

    Parameters:
    lst (list): nested list

    Returns:
    list: flattened list
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print("Deep flattened:", deep_flatten(nested_list))
```

8. Nested List Average

```python
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.

    Parameters:
    lst (list): nested list

    Returns:
    float: average value
    """
    flat_list = deep_flatten(lst)
    return sum(flat_list) / len(flat_list) if flat_list else 0

nested_list = [[1, 2], [3, 4], [5, 6]]
print("Average:", average_nested(nested_list))
```
